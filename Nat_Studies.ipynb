```python
df = df_player.merge(df_squad,on=['season','squad'],suffixes=("","_squad"))
```

```python
df = df[df['minutes_pl'] > 90]
```

```python
nats_each_year = []
years = []
nat_list_each_year = []
for year in df['season'].unique():
    years.append(year)
    one_season = df[df['season'] == year]
    list_nations = []
    nations = one_season['nationality'].apply(lambda x: str(x).split(",")).apply(pd.Series)
    #print(len(nations.columns))
    for col in nations.columns:
        list_nations.extend(nations[col].dropna().unique())
    unique = set(list_nations)
    if 'nan' in unique:
        unique.remove('nan')
    #print(list_nations)
    #print(unique)
    nat_list_each_year.append(unique)
    nats_each_year.append(len(unique))
    
```

```python
print(len(nat_list_each_year))
```

```python
plt.figure(figsize=(8,8))
plt.plot(years,nats_each_year,'r-',marker='^')
plt.xlabel('Year')
plt.ylabel('Nationalities Represented')
plt.title('EPL International participation by year')
plt.show()
```

```python
df['Pts'] = df['wins_squad'].apply(lambda x:x*3) + df['draws_squad']
temp_list = []
for year in df['season'].unique():
    one_season = df[df['season'] == year]
    for squad in one_season['squad'].unique():
        one_squad = one_season[one_season['squad'] == squad]
        list_nations = []
        nations = one_squad['nationality'].apply(lambda x: str(x).split(",")).apply(pd.Series)
        #print(len(nations.columns))
        for col in nations.columns:
            list_nations.extend(nations[col].dropna().unique())
        unique = set(list_nations)
        if 'nan' in unique:
            unique.remove('nan')
        temp_list.append({'season':year,'squad':squad,'squad_nat':len(unique)})
temp_df = pd.DataFrame(temp_list)
temp_df.head()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(y='season',x='squad_nat',data=temp_df,orient='h',order=(temp_df['season'].unique()[::-1]))
plt.ylabel('Year')
plt.xlabel('Nationalities Per Squad')
plt.title('EPL Squad Diversity by year')
plt.show()
```

```python
smaller = temp_df.merge(df[['Pts','squad','season','players_used']],on=['season','squad'])
```

```python
smaller = smaller.drop_duplicates()
smaller.head()
```

```python
smaller['fac_nat'] = smaller['squad_nat']/smaller['players_used']
```

```python
h = sns.jointplot(x='Pts',y='squad_nat',data=smaller,kind='kde')
h.set_axis_labels('League Points', 'Nationalities per Squad')
plt.show()
```

```python
mean_divers = smaller.groupby('season')['squad_nat'].mean()
```

```python
len(mean_divers)
```

```python
smaller = smaller.merge(mean_divers,on=['season'],suffixes=('','_mean'))
```

```python
smaller.head()
```

```python
smaller['rel_divers'] = smaller['squad_nat'] - smaller['squad_nat_mean']
```

```python
h = sns.jointplot(x='Pts',y='rel_divers',data=smaller,kind='kde')
h.set_axis_labels('League Points', 'Squad Relative Diversity')
plt.show()
```

```python
df_nat_counts = pd.DataFrame()
for year in df['season'].unique():
    years.append(year)
    one_season = df[df['season'] == year]
    nations = one_season['nationality'].apply(lambda x: str(x).split(",")).apply(pd.Series)
    #print(len(nations.columns))
    
    for col in nations.columns:
        df_counts = pd.DataFrame(nations[col].dropna().value_counts())
        df_counts['season'] = year
        df_counts['tot_players'] = len(one_season)
        df_nat_counts = df_nat_counts.append(df_counts)
df_nat_counts['num_players'] = df_nat_counts[0].fillna(0) + df_nat_counts[1].fillna(0)
```

```python
df_nat_counts = df_nat_counts.drop(0,axis=1)
df_nat_counts = df_nat_counts.drop(1,axis=1)
df_nat_counts.head()
```
